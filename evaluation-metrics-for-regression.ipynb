**Why do we need evaluation metrics?**

After training a model, we must answer one question: 
how good are the prediction?

evaluation metrics quantify prediction errors by comparing 
actual values y
predicted values -y

1 Mean absolute error MAE defi
mae is the avereage of absolute differences between actual and predicted values.

key properties 
easy to understand 
same unit as target variable
less sensitive to outliers


when to use:
when all errors are equally important

2. Root Mean Squared Error (RMSE) defi

RMSE is the square root of the average of squared errors

simple meaning

penalize larger error more strongly

example 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_true = np.array([50,60,70,80])
y_pred = np.array([48,62,68,85])

mae = mean_absolute_error(y_true, y_pred) 
mse = mean_squared_error(y_true, y_pred)

rmse = np.sqrt(mse)

r2 = r2_score(y_true, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2 Score:", r2)
```
